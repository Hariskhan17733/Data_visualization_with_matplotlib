### Task One: Creating data from an equation

It is important to be able to directly translate a real equation into a plot.
The [world famous equation](https://en.wikipedia.org/wiki/Mass%E2%80%93energy_equivalence) from Einstein:

$$E=mc^2$$

I used  Numpy to create two arrays: E and m , where **m** is simply 11 evenly spaced values representing 0 grams to 10 grams. E should be the equivalent energy for the mass.Assuming **c**=300000000

```python
import pandas as pf
import numpy as np
import matplotlib.pyplot as plt 
```

```python
m=np.linspace(0,10,11)
```

```python
m
```

```python
c=300000000
```

```python
e=m*c**2
```

```python
e
```

### Part Two: Plotting E=mc^2

Now that we have the arrays E and m, we can plot this to see the relationship between Energy and Mass. 



```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(m,e,color='red',lw=4)
ax.set_title('E MC^2')
ax.set_xlabel('Mass in Grams')
ax.set_ylabel('Energy in Joules')
ax.set_xlim(0,10);
```

## Task Two: Creating plots from data points

In finance, the yield curve is a curve showing several yields to maturity or interest rates across different contract lengths (2 month, 2 year, 20 year, etc. ...) for a similar debt contract. The curve shows the relation between the (level of the) interest rate (or cost of borrowing) and the time to maturity, known as the "term", of the debt for a given borrower in a given currency.

The U.S. dollar interest rates paid on U.S. Treasury securities for various maturities are closely watched by many traders, and are commonly plotted on a graph such as the one on the right, which is informally called "the yield curve".



## Part One: Yield Curve Data

**We've obtained some yeild curve data for you from the [US Treasury Dept.](https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/textview.aspx?data=yield). The data shows the interest paid for a US Treasury bond for a certain contract length. The labels list shows the corresponding contract length per index position.**



```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

**Figure out how to plot both curves on the same Figure. Add a legend to show which curve corresponds to a certain year.**

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(labels,july16_2007,label='july16_2007',color='orange',lw=1)
ax.plot(labels,july16_2020,color='blue',lw=1,label='july16_2020')
ax.legend(loc=(1.1,0.5))
;
```

**While the plot above clearly shows how rates fell from 2007 to 2020, putting these on the same plot makes it difficult to discern the rate differences within the same year.**

```python
fig,axes=plt.subplots(2,1)
axes[0].plot(labels,july16_2007,label='july16_2007',lw=1)
axes[0].set_title('july16_2007')
axes[1].plot(labels,july16_2020,label='july16_2020',lw=1)
axes[1].set_title('july16_2020')
#fig.tight_layout()
fig.subplots_adjust(wspace=0.2,hspace=0.3)
```

**Recreate the plot below that uses twin axes.**

```python
fig,ax1=plt.subplots()
ax1.plot(labels,july16_2007,label='july16_2007',lw=1,color='blue')
ax1.set_ylabel(r"20007", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
ax1.spines['left'].set_color('blue')
ax1.spines['left'].set_linewidth(4)
ax2 = ax1.twinx()
#for second
ax2.plot(labels,july16_2020,label='july16_2020',lw=1,color='red')
ax2.set_ylabel(r"2020", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
ax1.spines['right'].set_color('red')
ax1.spines['right'].set_linewidth(4)

;
```
